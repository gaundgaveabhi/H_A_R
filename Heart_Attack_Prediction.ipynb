# Missing Data - Initial Intuition

* Here, we don't have any missing data.

General Thumb Rules:

* For features with less missing values- can use regression to predict the missing values or fill with the mean of the values present, depending on the feature.
* For features with very high number of missing values- it is better to drop those columns as they give very less insight on analysis.
* As there's no thumb rule on what criteria do we delete the columns with high number of missing values, but generally you can delete the columns, if you have more than 30-40% of missing values. But again there's a catch here, for example, Is_Car & Car_Type, People having no cars, will obviously have Car_Type as NaN (null), but that doesn't make this column useless, so decisions has to be taken wisely.

# LabelEncoding 

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

data=df.select_dtypes('object')
dict1={}
for column in data:
    dict1[column]=label
    data[column]=dict1[column].fit_transform(df[column])
data.head() 
```

```python
df.drop(['Sex','Diet','Hemisphere'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df2=pd.concat([data,df],axis=1)
```

```python
df2.head()
```

```python
df2.info()
```

# Handling Inbalanced datassets:

```python
df['Heart Attack Risk'].value_counts()
```

```python
x=df2.drop('Heart Attack Risk',axis=1)
x.shape
```

```python
y=df2['Heart Attack Risk']
y.shape
```

###### As you can see that the accuracy is quite low, and as it's an imbalanced dataset, we shouldn't consider Accuracy as our metrics to measure the model, as Accuracy is cursed in imbalanced datasets.

###### Hence, we need to check recall, precision & f1 score for the minority class, and it's quite evident that the precision, recall & f1 score is too low for Class 1, i.e. churned customers.

###### Hence, moving ahead to call SMOTEENN (UpSampling + ENN)

```python
from imblearn.combine import SMOTEENN 
sme = SMOTEENN(random_state=42)
X_resampal, y_resampal = sme.fit_resample(x, y)
```

```python
X_resampal.shape
```

```python
y_resampal.shape
```

# Model building:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

# Model train_test_split

```python
xtrain,xtest,ytrain,ytest= train_test_split(X_resampal,y_resampal,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

# LogisticRegression model

```python
from sklearn.linear_model import LogisticRegression
```

# Model building

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

# Model evaluation

```python
# model pridection test data
y_pred=model.predict(xtest)

#accuracy_score
print("accuracy_score in test data",accuracy_score(ytest,y_pred))

#classcification report
print('classification_report in test data :\n\n',classification_report(ytest,y_pred))

#confusion_matrix
print('Confusion Matrix \n',confusion_matrix(ytest,y_pred))
```

```python
#Model pridection train data
x_pred=model.predict(xtrain)

#accuracy_score
print('accuracy_score in train data' ,accuracy_score(ytrain,x_pred))

#classification_report
print('classification_report :\n\n',classification_report(ytrain,x_pred))

#confusion_matrix
print('Confusion Matrix \n',confusion_matrix(ytrain,x_pred))
```

Now we can see quite better results, i.e. Accuracy: 70 %, and a very good recall, precision & f1 score for minority class.
